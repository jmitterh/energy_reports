```python
import gc
def os_walk_to_df():
    num = 1
    for root, dirs, files in os.walk('./data'):
        for file in files:
            # dataframe
            try:
                df = pd.read_csv(os.path.join(root, file))
            except UnicodeDecodeError:
                print(f'Error reading {file} due to encoding, trying ISO-8859-1\n')
                df = pd.read_csv(os.path.join(root, file), encoding='ISO-8859-1')
            print(f'File {num} is {file}')
            print(df.head())
            print(df.info())
            print(df.describe())
            print(df.columns)
            # check if datetime is in cronological order
            print(df['Timestamp'].head())
            print(df['Timestamp'].tail())
            print(f'\n___________________________________________________________\n')
            # deleting dataframe for next iteration
            del df
            gc.collect()
            num += 1
    return
            
```

### Aggregate 30 minute intervals into 1 hour intervals
- Since this is operational monitoring data, we can assume that the data is continuous and that there are no missing values. Therefore, we can aggregate the data into 1 hour intervals to see how power output fluctuates on average.

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp']) 
df = df.set_index('Timestamp') 
df = df.resample('h').agg({'Power (MW)': 'mean','Press (psig)': 'mean', 'Temp (°F)': 'mean', 'PowerSwing (MW)': 'mean'})
df = df.reset_index() 
df.head()
```

### Calculate supporting columns for Enthalpy and add it as new columns
- add any supporting columns if needed

```python
from  iapws import IAPWS97
# Pounds per square gauge (PSIG) to pounds per square inch absolute (PSIA or total pressure)
# https://www.thecompressedairblog.com/psi-psia-and-psig-whats-the-difference#:~:text=PSIA%20is%20pounds%20per%20square,relative%20to%20ambient%20atmospheric%20pressure.
df['Press_psia'] = df['Press (psig)'] + 14.7
# Convert the temperature from Farenheit to Kelvin
df['temp_degK'] = (df['Temp (°F)'] - 32) * 5/9 + 273.15
# Convert PSIA to megapascals (MPa)
df['Press_MPa'] = df['Press_psia'] * 6894.76 / 10**6 # 1,000,000 (1 psi = 6894.76 Pa)
df.head()
```

```python
# h: Specific enthalpy, [kJ/kg] = IAPWS97(mega_pascals, Temperature_Kelvin)
def calculate_enthalpy(row):
    try:
        return IAPWS97(P=row['Press_MPa'], T=row['temp_degK']).h
    except NotImplementedError as e: # temp., I believe, does meet the IAPWS97 standard
        print(f"Error for Timestamp {row['Timestamp']} P={row['Press_MPa']}, T={row['temp_degK']}: {e}")
        return None

df['Steam'] = df.apply(calculate_enthalpy, axis=1)
```
